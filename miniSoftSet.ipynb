```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
data = {'świeże':[0.9,0.8,1,0.6,0.5,0.9,1],
       'mrożone':[1,0,0,0,0,0,0],
       'ostre':[0,0.2,0.1,0.1,0.2,0,0.3],
       'słodkie':[0.8,0.8,0.7,0.4,0.5,0.2,0.1],
       'zielone':[0,0,0,0,1,0,0],
       'czerwone':[1,1,1,1,0,0,1],
       'lokalne':[0.4,0.8,0,0,1,1,0.9],
       'tropikalne':[0.6,0.2,1,1,0,0,0.1],
       'liściaste':[0,0,0,0,0,0,0],
       'bulwowe':[0,0,0,0,0,1,1]}

df = pd.DataFrame(data=data, index = {'maliny mrożone',
                                      'pomidory malinowe',
                                      'pomidory koktajlowe',
                                      'papryka czerwona',
                                      'papryka zielona',
                                      'ziemniaki sałatkowe',
                                      'buraki czerwone'})

A = {'świeże': 1, 'ostre': 0.8, 'czerwone': 0.5}
B = {'mrożone': 1, 'zielone': 0.8, 'słodkie': 0.5, 'liściaste': 0.2}
C = {'świeże': 1, 'zielone': 0.5, 'czerwone': 0.5, 'słodkie': 0.9}

df
```

```python
class SoftSet:
    @staticmethod
    def bestInference(u: pd.DataFrame, a: dict):
        dic = defaultdict(lambda: 0,a)
        e = u.columns.tolist()
        result = {}
        for i in u.index:
            sumx = 0
            for parameter in e:
                sumx += u.at[i,parameter] * dic[parameter]
            result[i] = sumx
        result = sorted(result.items(), key = lambda x:x[1], reverse = True)
        return result[0:3]
```

```python
test = SoftSet.bestInference(df,A)
print(f"Dla A: {test}")
print()

test1 = SoftSet.bestInference(df,B)
print(f"Dla B: {test1}")
print()

test2 = SoftSet.bestInference(df,C)
print(f"Dla C: {test2}")
```
