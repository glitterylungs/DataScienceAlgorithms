```python
import pandas as pd
import random 
import math
import numpy as np
```

```python
class ProcessingData:
    @staticmethod
    def shuffleDF(data:pd.DataFrame)->pd.DataFrame:
        for i in range(len(data)-1,0,-1):
            x=random.randint(0,i)
            data.iloc[i], data.iloc[x] = data.iloc[x], data.iloc[i]
        return data
    
    @staticmethod
    def normalizeDF(data:pd.DataFrame)->pd.DataFrame:
        col=data.columns.tolist()
        col=col[:-1]
        for j in col:
            minn=data[j].min()
            maxx=data[j].max()
            for i in range(0,len(data),1):
                data.at[i,j]=(float(data.at[i,j])- minn)/(maxx - minn)
        return data
    
    @staticmethod
    def splitDF(data:pd.DataFrame)->pd.DataFrame:
        trainDF=pd.DataFrame(columns=data.columns)
        testDF=pd.DataFrame(columns=data.columns)
        for i in range(len(data)):
            if i <= len(data)*0.7:
                trainDF=trainDF.append(data.iloc[i])
            else:
                testDF=testDF.append(data.iloc[i])
        return trainDF, testDF
```

```python
class KNN:
    @staticmethod
    def dist(x: list,y: list,m: int):
        result = 0
        n = len(x)
        for i in range(0,n-1,1):
            result += (abs(x[i]-y[i]))**m
        minkowsky = result**(1/m)
        return minkowsky
    
    @staticmethod
    def q(sample: list, data: pd.DataFrame, m: int, k: int):
        distances = []
        for i in range(0,len(data)):
            row=data.iloc[i].tolist()
            distances.append([KNN.dist(row[:-1],sample,m),row[-1]])
        distances.sort(key = lambda distance:distance[0])
        types = {}
        for v in pd.unique(data["variety"]):
            types[v] = 0
        for l in range(k):
            types[distances[l][1]] += 1
        return max(types, key=types.get)
    
    @staticmethod
    def precision(testData: pd.DataFrame, trainData: pd.DataFrame, m: int, k:int):
        good = 0
        bad = 0
        for data in testData.values:
            if KNN.q(data,trainData,m,k) == data[-1]:
                good += 1
            else:
                bad += 1
        return good/(bad+good) * 100
            
    
```
